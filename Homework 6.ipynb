```python
from matplotlib import pyplot as plt
log = open('HW6_scripts/log_lenet.txt', 'r')
train, test = [], []
for line in log:
    if line[1]!='*':
        continue
    if line[3:16]=='Training Prec':
        train.append(float(line[19:-1]))
    elif line[3:15]=='Testing Prec':
        test.append(float(line[18:-1]))
log.close()
plt.figure(figsize=(20,10))
plt.plot(train, label='Training Accuracy')
plt.plot(test, label='Testing Accuracy')
plt.title('LeNet Training/Testing Accuracy')
plt.legend()
plt.show()
```
